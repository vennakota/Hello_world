```python
import os
import pandas as pd 
import seaborn as sns
```

```python
titanic= pd.read_csv("Titanic1.csv", sep=",")
```

```python
sns.countplot(x='Survived',data=titanic) #seaborn is used to analyise how the project the data
```

```python
sns.catplot(x='Pclass',data=titanic, palette= 'Set2',col='Survived', kind='count') #barplot
```

```python
sns.catplot(x='Pclass',data=titanic, palette= 'Set2',hue='Sex',col='Survived', kind='count')# it is telling that to dividethe data with sex 
```

#hue is a categorical variable so don't try to keep the numeric variable

```python
titanic['age_group']=['<13'if x<=13 else'13-18' if x <=18 else'18-40' if x<=40 else'40-60' if x<=60 else'60-80' if x<=80 else '80+'if x>80 else'Not_given' for x in titanic['Age']]
```

```python
sns.catplot(x='Age',data=titanic, palette= 'Set1',col='age_group', kind='strip')
```

```python
# if you keep strip we will see the dots, if you keep the count it just give a normal plot, 
```
